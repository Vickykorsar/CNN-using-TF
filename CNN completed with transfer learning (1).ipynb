```python
from keras.layers import Dense,Lambda,Input, Flatten
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.models import Sequential
```

```python
vgg=VGG16(input_shape=[130,130,3],
         weights='imagenet',
         include_top=False)
```

```python
for i in vgg.layers:
    i.trainable=False
```

```python
x=Flatten()(vgg.output)
```

```python
#y=Dense(20,activation='relu')(x)
```

```python
z=Dense(1,activation='sigmoid')(x)
```

```python
model=Model(vgg.input,z)
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(F,T,epochs=10,validation_split=0.15)
```
